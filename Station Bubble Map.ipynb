```python
#Vizualization of Popular Train Stations Using Bubble Map

import numpy as np
import folium
import pandas as pd
import branca
amtrackData = pd.read_csv("amtrackData.csv")

station_popularity = amtrackData.groupby('STNNAME').agg({
    'Train ID': 'nunique',
    'X': 'first',
    'Y': 'first'
}).reset_index()

station_popularity.rename(columns={'Train ID': 'Train Count'}, inplace=True)
station_popularity['Log Train Count'] = np.log1p(station_popularity['Train Count'])
m = folium.Map(location=[37.0902, -95.7129], zoom_start=4, tiles="OpenStreetMap")

min_val = station_popularity['Log Train Count'].min()
max_val = station_popularity['Log Train Count'].max()
mid_val = (min_val + max_val) / 2


colormap = branca.colormap.LinearColormap(
    colors=['lightblue', 'darkblue'],
    vmin=min_val,
    vmax=max_val,
    caption="Station Popularity"
)
colormap.add_to(m)


scale_factor = 3

for index, row in station_popularity.iterrows():
    folium.CircleMarker(
        location=(row['Y'], row['X']),
        radius=row['Log Train Count'] * scale_factor,
        color='black',  
        weight=1,  
        fill=True,
        fill_color=colormap(row['Log Train Count']),
        fill_opacity=0.7,
        tooltip=f"Station: {row['STNNAME']}, Frequency: {row['Train Count']}"
    ).add_to(m)

m
```
